```python
#B

def troca_de_colunas (df, coluna1, coluna2):

    colunas = list (df)
    colA = colunas.index (coluna1)
    colB = colunas.index (coluna2)
    colunas [colA] = coluna2
    colunas [colB] = coluna1
    df.reindex (columns = colunas)
    
    return df.reindex (columns = colunas)

df_8 = pd.DataFrame(np.arange(20).reshape(-1, 5), columns=list('abcde'))

df_8 = troca_de_colunas (df_8, 'a', 'e') #só modificar aqui as letras
df_8
```
