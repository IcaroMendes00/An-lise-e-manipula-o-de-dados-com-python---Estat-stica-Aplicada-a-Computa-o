```python
#1

Mercearia = pd.DataFrame({"loja":["A", "A", "A"], "dia": ["sex", "sab", "dom"], "valor": [7500, 9500, 8200]})

Farmácia = pd.DataFrame({"loja":["B", "B", "B"], "dia": ["sex", "sab", "dom"], "valor": [5100, 8250, 9900]})

Restaurante = pd.DataFrame({"loja":["C", "C"], "dia": ["sab", "dom"], "valor": [7500, 11800]})
```

```python
pd.concat([Mercearia, Farmácia, Restaurante], axis=1)
```

```python
pd.concat([Mercearia, Farmácia, Restaurante], ignore_index=True)
```

A diferença se da porque da primeira forma, o axis=1 informa em qual sentido vai se dar a concatenação, no caso sentido horizontal, combinando por linha, as linhas 0 uma ao lado da outra, aslinhas 1 uma ao lado das outra..., então ao formar o data frame os valores da linha 2 foram preenchidos com um valor nulo, já que o data frame Restaurante não possuia uma linha 2, apenas a linha 0 e 1.

Já o segundo data frame formado possui um melhor formato, já que os ordenou por colunas, coluna x abaixo de coluna x, coluna y abaixo de coluna y..., ignore_index=True faz com que os índices sejam reordenados, ou seja, ao invés de se repetirem, serão modificados de acordo com sua ordem.  

```python
#2

SQL = pd.DataFrame({"email":["rakesh@xyz.com", "ecg@acmecorpus.com"]})

PYTHON = pd.DataFrame({"email":["ana@xyz.com", "jonas@acmecorpus.com", "rakesh@xyz.com"]})
```

```python
#(a) quais são os alunos distintos, considerando ambas as listas?

Distintos = pd.concat([SQL, PYTHON],ignore_index=True)
Distintos = Distintos.drop_duplicates()
Distintos
```

```python
#(b) quais são os alunos que fizeram ambos os cursos?

Ambos = SQL.merge(PYTHON)
Ambos
```

```python
#3

dic_emp = {"id":["D1","D2","D3","D4"], 
           "nomDepto": ["Compras","RH","TI","Vendas"], 
           "local":["SP","RJ","RJ","SP"] }

dic_func = {"num":[3199,3269,3555,3788,3844], 
            "nome": ["Ana","David","José","Marina","Luís"], 
            "salario":[1600,2975,1500,5000,3000], 
            "idDpto": ["D2","D3",None,"D2","D4"] }

empresa = pd.DataFrame(dic_emp)

funcionarios = pd.DataFrame(dic_func)

df_geral = pd.merge(funcionarios, empresa, how="left", left_on="idDpto", right_on="id")
df_geral
```
