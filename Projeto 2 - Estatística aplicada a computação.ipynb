#### 2 - Letra C

Fiz um df copia e filtrei as colunas ia precisar para não modificar o df original. Modifiquei a coluna de "data_inversa" para deixar só com o ano nos dois dataframes

```python
df2019_cop = df2019.filter(items= ['data_inversa',"uf",'dia_semana',"municipio"])
df2020_cop = df2020.filter(items= ['data_inversa',"uf",'dia_semana',"municipio"])
df2019_cop["data_inversa"] = 2019
df2020_cop["data_inversa"] = 2020
```

Depois usando o comando .groupby() mais o .size() para agrupar. O comando .sort_values(ascending=False) faz com que esses numeros somado fiquem em ordem decrescente, assim com o .head(5) peguei as 5 primeiras linhas. ao usar o comando groupby ele deixa o df com MultiIndex, para evitar isso usei .reset_index() que remove um ou mais niveis. 

```python
mais_acid_2019 = df2019_cop.groupby(["data_inversa","municipio"]).size().sort_values(ascending=False).head(5).reset_index()
mais_acid_2020 = df2020_cop.groupby(["data_inversa","municipio"]).size().sort_values(ascending=False).head(5).reset_index()
```

Agora só para deixar o dataframe mais compreensível mudei os nomes das colunas e juntei os dois em um só, usando o comando concat e ,axis = 1 para que a união fosse no sentido das linhas

```python
mais_acid_2019.rename(columns = {0:"quant_acidentes", "data_inversa":"Ano"}, inplace = True)
mais_acid_2020.rename(columns = {0:"quant_acidentes", "data_inversa":"Ano"}, inplace = True)
mais_acid_final = pd.concat([mais_acid_2019,mais_acid_2020])
```

As cincos cidades com mais acidentes em 2019 e 2020 foram:

```python
mais_acid_final.set_index(['Ano',"municipio","quant_acidentes"]) #para deixar mais organizado fiz com MultiIndex
```

Primeiro fiz um dataframe para cada ano, filtrando as colunas necessárias para fazer a questão

#### (b) qual a proporção de pessoas ilesas e de feridos graves por mês em cada um dos anos considerados? (proporção = numero de ilesos ou feridos graves/ número total de pessoas envolvidas no acidente);

###### 4 - letra A
fiz uma serie para saber o tatal de acidentes pela coluna "causa_acidente", usando .groupy() e size()

```python
causas2019 = df2019.groupby("causa_acidente").size()
causas2020 = df2020.groupby("causa_acidente").size()
```

Fiz uma Função para saber se tem alguma causa com que tenha o mesmo numero de acidente

```python
def causas_raras(lista):
    if (min(lista) == 1):
        temp = lista == 1 #aqui é para saber se alguma causa tem o valor 1
        causas_raras = lista[temp] #aqui filtra a lista com as cusas raras
        print("Os acidentes mais raros foram: ")
        return causas_raras

    else:
        print('A causa de acidentes com mais rara foi, {}, com {} acidentes'.format(lista.idxmin(),lista.min()))
        #usando .idxmin() e .min() para retornar o rotulo e menor valor
```

Para saber a mais frequente usei a função .idxmax() que retorna o rótulo com maior valor, e .max() que retorna o maior valor da lista  

```python
print("-------2019-------")
print()
print('A causa de acidentes com mais frequencia foi, {}, com {} acidentes'.format(causas2019.idxmax(),causas2019.max()))
print()
causas_raras(causas2019)
```

```python
print("-------2020-------")
print()
print('A causa de acidentes com mais frequencia foi {} com {} acidentes'.format(causas2019.idxmax(),causas2019.max()))
print()
causas_raras(causas2020)
```

###### 4 - letra B
Primeiro juntei os dois dataframes usando .concat() e em seguinda, filtrei as colunas que interessa para questão

```python
df_junto = pd.concat([df2019,df2020])
df_selecionado= df_junto.filter(items = ['data_inversa','feridos_graves','ilesos','pessoas'])
```

depois usando .astype('datetime64'), na coluna 'data_inversa' para garantir que a coluna fique no formatato de data, para ser usada a função .map(lambda x: 100*x.year + x.month), que tranforma (2019-01-01) em 201901, para poder agrupar.

```python
df_selecionado['data_inversa'] = df_selecionado['data_inversa'].astype('datetime64')
df_selecionado['Ano e Mes'] = df_selecionado['data_inversa'].map(lambda x: 100*x.year + x.month)
df_selecionado
```

- se executar a célula mais de uma vez vai dar erro, porque o codigo já alterou a data, se acontecer é só executar a célula anterior

Em seguinda fiz a soma da coluna "Ano e Mes" para poder organizar o dataframe

```python
df_selecionado = df_selecionado.groupby(['Ano e Mes']).sum().reset_index()
df_selecionado
```

e para finalizar, fiz a soma do que se pede na questão e assim tendo a porporção.

```python
df_selecionado["Proporção"] = (df_selecionado['feridos_graves'] + df_selecionado['ilesos']) / df_selecionado['pessoas']
df_selecionado
```

###### 4 - letra C

nessa confesso que sofri bastante...

Aqui é uma lista que vai ser usada para deixar o codigo mais organizado

```python
meses = ["Janeiro","Fevereiro","Março","Abril","Maio","Junho","Julho","Agosto","Setembro","Outubro","Novembro","Dezembro"]
```

Em seguida, fiz um novo dataframe de acada ano só com as colunas "data_inversa" e "br" para poder calcular o total de acidentes por br.

```python
df_br_2019 = df2019.filter(items = ['data_inversa','br'])
df_br_2019['data_inversa'] = df_br_2019['data_inversa'].astype('datetime64')
df_br_2019 = df_br_2019.groupby(['data_inversa',"br"]).size().reset_index()
df_br_2019
```

Depois usando o comando .map(lambda x: + x.month) para colocar numero por mês e depois usando para somar e ter o mes e o tatal de acidentes por br

```python
df_br_2019['data_inversa'] = df_br_2019['data_inversa'].map(lambda x: + x.month)
df_br_2019.rename(columns= {0:"Acidedentes_2019"}, inplace = True)
df_br_2019 = df_br_2019.groupby(['data_inversa']).sum().reset_index()
df_br_2019
```

- Um pequeno detalhe é que ao somar o mês o dataframe ele também somou as brs, mas ela não vai ser mais usada aparti daqui
- se executar a célula mais de uma vez vai dar erro, porque o codigo já alterou a data, se acontecer é só executar a célula anterior

Fiz a mesma coisa com 2020

```python
df_br_2020 = df2020.filter(items = ['data_inversa','br'])
df_br_2020['data_inversa'] = df_br_2020['data_inversa'].astype('datetime64')
df_br_2020 = df_br_2020.groupby(['data_inversa',"br"]).size().reset_index()
df_br_2020
```

```python
df_br_2020['data_inversa'] = df_br_2020['data_inversa'].map(lambda x: + x.month)
df_br_2020.rename(columns= {0:"Acidedentes_2020"}, inplace = True)
df_br_2020 = df_br_2020.groupby(['data_inversa']).sum().reset_index()
df_br_2020
```

Aqui troquei os nomeros dos meses pelo nome usando a lista já definida anteriormente 

```python
df_br_2019['data_inversa'] = meses
df_br_2020['data_inversa'] = meses
```

Usando o comando .merge que combina dataframed aparti de colunas ou linhas repetidas. Com isso juntei os dois dataframes baseado na coluna "data_inversa"

```python
pd_temp = pd.merge(df_br_2019, df_br_2020, on="data_inversa")
pd_temp
```
